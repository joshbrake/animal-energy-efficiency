```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('animal-energy-efficiency.csv')
```

```python
fig, ax = plt.subplots(dpi=200)

ax.plot(df['x'],df['y'],'o', ms = 3)
ax.set_xscale('log')
ax.set_yscale('log')

yoffset = 1.1
for i in range(len(df)):
    ax.text(x=df['x'][i], y=df['y'][i]*yoffset, s = df['Description'][i], 
            verticalalignment='bottom', horizontalalignment='center',
           size='x-small')

ax.set_xlabel('Body Mass [kg]')
ax.set_ylabel('Cost of Transport [kcal/kg/km]')
ax.grid(which='major', ls='-')
ax.grid(which='minor', ls='dotted')
```

```python
fig.savefig('animal-energy-efficiency.pdf')
```
